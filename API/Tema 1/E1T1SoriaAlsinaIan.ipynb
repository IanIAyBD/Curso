```python
def es_primo(num):
    primo = True
    if num == 2:
        return primo
    else:
        for i in range(2,num):
            if (num%i) == 0:
                primo = False
        return primo

primos = []
for i in range(2,100):
    if es_primo(i) == True:
        primos.append(i)
print(primos)
    
```

```python
for i in range(0,len(primos)):
    print(i+1,"º -> ",primos[i])
```

```python
for i in range(0,len(primos)):
    if (i == 0):
        print("Primero ->",primos[i])
    elif (i == len(primos)-1):
        print("Ultimo ->",primos[i])
    else:
        print(i+1,"º -> ",primos[i])
```
