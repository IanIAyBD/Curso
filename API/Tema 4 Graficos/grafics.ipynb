#### Generar los valores de muestra (inventados):

```python
import numpy as np
# Funcion ejemplo para generar datos, combinaci√≥n simple de una funci√≥n lineal y sinusoidal.
# Definimos la funci√≥n con lambda.
valores = lambda A, amp, angle, phase: A * angle + amp * np.sin(angle + phase)

# Define parameters 
theta = np.linspace(0., 2 * np.pi, 250) # x-axis (x ir√° desde 0 a 2pi)
np.random.seed(100)  #inicializamos semilla (para el random)
noise = 0.2 * np.random.random(250)  #generamos ruido (aleatorio)
y = valores(.1, .2, theta, 1.2) + noise # y-axis
```

#### Ahora todo junto:

```python
from matplotlib.patches import ConnectionPatch
import matplotlib.pyplot as plt
import numpy as np

# Funcion ejemplo para generar datos, combinaci√≥n simple de una funci√≥n lineal y sinusoidal.
# Definimos la funci√≥n con lambda.
valores = lambda A, amp, angle, phase: A * angle + amp * np.sin(angle + phase)

# Define parameters 
theta = np.linspace(0., 2 * np.pi, 250) # x-axis (x ir√° desde 0 a 2pi)
np.random.seed(100)  #inicializamos semilla (para el random)
noise = 0.2 * np.random.random(250)  #generamos ruido (aleatorio)
y = valores(.1, .2, theta, 1.2) + noise # y-axis

fig = plt.figure(figsize=(6, 5))
plt.subplots_adjust(bottom = 0., left = 0., top = 1., right = 1.) #ajustamos tama√±o. Tambi√©n se le puede pasar wspace y hspace separaci√≥n entre axes

sub1 = fig.add_subplot(2,2,1) # dos filas, dos columnas, primera celda
#En el primer axes, arriba a la izquierda, gr√°fico en verde.
sub1.plot(theta, y, color = 'green')
#establecemos los l√≠mites.
sub1.set_xlim(1, 2)
sub1.set_ylim(0.2, .5)
#etiqueta
sub1.set_ylabel('y', labelpad = 15)

sub2 = fig.add_subplot(2,2,2) # dos filas, dos columnas, segunda celda
#En el segundo axes, arriba a la derecha, gr√°fico en naranja.
sub2.plot(theta, y, color = 'orange')
sub2.set_xlim(5, 6)
sub2.set_ylim(.4, 1)

sub3 = fig.add_subplot(2,2,(3,4)) # dos filas, dos columnas, combinada tercera y cuarta celda
#En el tercer axes, combina tercera y cuarta celda.
sub3.plot(theta, y, color = 'darkorchid', alpha = .7)
sub3.set_xlim(0, 6.5)
sub3.set_ylim(0, 1)
sub3.set_xlabel(r'$\theta$ (rad)', labelpad = 15) #etiquetas y separaci√≥n
sub3.set_ylabel('y', labelpad = 15)  #etiquetas y separaci√≥n

#Crear una area bloqueada (fill_between)
sub3.fill_between((1,2), 0, 1, facecolor='green', alpha=0.2) # blocked area for first axes
sub3.fill_between((5,6), 0, 1, facecolor='orange', alpha=0.2) # blocked area for second


# ConnectionPatch une incluso areas de diferentes axes (gr√°ficos) o zonas en la figura (figure)
con1 = ConnectionPatch(xyA=(1, .2), coordsA=sub1.transData, 
                       xyB=(1, .3), coordsB=sub3.transData, color = 'green')
# Add left side to the figure
fig.add_artist(con1)

# Create right side of Connection patch for first axes
con2 = ConnectionPatch(xyA=(2, .2), coordsA=sub1.transData, 
                       xyB=(2, .3), coordsB=sub3.transData, color = 'green')
# Add right side to the figure
fig.add_artist(con2)

# Create left side of Connection patch for second axes
con3 = ConnectionPatch(xyA=(5, .4), coordsA=sub2.transData, 
                       xyB=(5, .5), coordsB=sub3.transData, color = 'orange')
# Add left side to the figure
fig.add_artist(con3)

# Create right side of Connection patch for second axes
con4 = ConnectionPatch(xyA=(6, .4), coordsA=sub2.transData, 
                       xyB=(6, .9), coordsB=sub3.transData, color = 'orange')
# Add right side to the figure
fig.add_artist(con4)

# Save figure with nice margin
plt.savefig('zoom_effect_3.png', dpi = 300, bbox_inches = 'tight', pad_inches = .1)
```

#### Ejemplo ConnectionPatch

```python
from matplotlib.patches import ConnectionPatch
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 3))

# Draw a simple arrow between two points in axes coordinates
# within a single axes.
xyA = (0.2, 0.2)
xyB = (0.8, 0.8)
coordsA = "data"
coordsB = "data"
con = ConnectionPatch(xyA, xyB, coordsA, coordsB,
                      arrowstyle="-|>", shrinkA=5, shrinkB=5,
                      mutation_scale=20, fc="w")
ax1.plot([xyA[0], xyB[0]], [xyA[1], xyB[1]], "o")
ax1.add_artist(con)

# Draw an arrow between the same point in data coordinates,
# but in different axes.
xy = (0.3, 0.2)
coordsA = "data"
coordsB = "data"
con = ConnectionPatch(xyA=xy, xyB=xy, coordsA=coordsA, coordsB=coordsB,
                      axesA=ax2, axesB=ax1,
                      arrowstyle="->", shrinkB=5)
ax2.add_artist(con)

# Draw a line between the different points, defined in different coordinate
# systems.
xyA = (0.6, 1.0)  # in axes coordinates
xyB = (0.0, 0.2)  # x in axes coordinates, y in data coordinates
coordsA = "axes fraction"
coordsB = ax2.get_yaxis_transform()
con = ConnectionPatch(xyA=xyA, xyB=xyB, coordsA=coordsA, coordsB=coordsB,
                      arrowstyle="-")
ax2.add_artist(con)

ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax2.set_xlim(0, .5)
ax2.set_ylim(0, .5)

plt.show()
```

```python
#Al dibujar una superficie en 3D, podemos cambiar los colores en funci√≥n del valor de z. Para ello usamo el argumento cmap.
import matplotlib.pyplot as plt
import numpy as np
  
figure=plt.figure(figsize=(8,8))
axes = figure.add_subplot(projection='3d')
 
x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
x,y=np.meshgrid(x,y)
z =  3*(1 - x)**2 * np.exp(-x**2 - (y + 1)**2)  - 10*(x/5 - x**3 - y**5)*np.exp(-x**2 - y**2) - 1./3*np.exp(-(x + 1)**2 - y**2) 
 
axes.plot_surface(x,y,z, cmap='turbo')
```

```python
#Podemos a√±adir una barra con los colores con el m√©todo colorbar.
#Notar que el m√©todo colorbar se usa sobre la figura y hay que pasarle el axes.
#Por √∫ltimo el par√°metro shrink es para que la barra no salga tan alta.
import matplotlib.pyplot as plt
import numpy as np
  
figure=plt.figure(figsize=(8,8))
axes = figure.add_subplot(projection='3d')
 
x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
x,y=np.meshgrid(x,y)
z =  3*(1 - x)**2 * np.exp(-x**2 - (y + 1)**2)  - 10*(x/5 - x**3 - y**5)*np.exp(-x**2 - y**2) - 1./3*np.exp(-(x + 1)**2 - y**2) 
 
surface=axes.plot_surface(x,y,z,cmap='turbo')
figure.colorbar(surface,ax=axes, shrink = 0.5)

#Tambi√©n se puede hacer:
#figure.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.turbo),ax=axes, shrink = 0.5)
```

```python
#Se pueden a√±adir sombreado a la imagen para que quede mas realista.
import matplotlib.pyplot as plt
from matplotlib.colors import LightSource
import numpy as np
  
figure=plt.figure(figsize=(8,8))
axes = figure.add_subplot(projection='3d')
 
x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
x,y=np.meshgrid(x,y)
z =  3*(1 - x)**2 * np.exp(-x**2 - (y + 1)**2)  - 10*(x/5 - x**3 - y**5)*np.exp(-x**2 - y**2) - 1./3*np.exp(-(x + 1)**2 - y**2) 
 
light_source = LightSource()
facecolors = light_source.shade(z,plt.cm.turbo, blend_mode='soft')
 
axes.plot_surface(x,y,z,facecolors=facecolors)
```
