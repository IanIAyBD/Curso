### 3 PREGUNTAS:

#### ¿Cuántos jugadores están en el FC Barcelona?

##### <<Solución>> (33)

```python
df[df['Club'] == 'FC Barcelona']
```

#### ¿Cuántos jugadores son de España?

##### <<Solución>> (1072)

```python
df[df['Nationality'] == 'Spain']
```

#### ¿Cuál es la puntuación máxima que se le da en FIFA?   (Overall)

##### <<Solución>> (94. Son 2, Messi y Ronaldo)

```python
df[df['Overall'] == df['Overall'].max()]
```
