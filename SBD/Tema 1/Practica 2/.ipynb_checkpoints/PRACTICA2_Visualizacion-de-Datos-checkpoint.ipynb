Leed el archivo csv utilizando la librería solicitada.


```python
ratings = "teachingratings.csv"
ratings_df = pd.read_csv(ratings)
```

## Ejercicios de Laboratorio


Utilizando la variable prof, encontrad los valores únicos de dicha variable.


```python
ratings_df['prof'].unique()
```

Utilizando toda las observaciones, mostrad el promedio para la edad.


```python
ratings_df['age'].mean()
```

Repetid el análisis filtrando primero el conjunto de datos para incluir una observación por cada instructor con un número total de observaciones = 94.

>Primero descartamos los duplicados utilizando la variable prof como sub conjunto y asignarle un nuevo dataframe llamado no_duplicates_ratings_df


```python
no_duplicates_ratings_df = ratings_df.drop_duplicates(subset=['prof'])
no_duplicates_ratings_df.head(20)
```

>Utilizamos el nuevo conjunto de datos para obtener la media de edad.


```python
no_duplicates_ratings_df['age'].mean()
```

Mediante una gráfica de barras (barplot de la librería seaborn), demostrad si los cursos de niveles inferiores (variable division) obtienen en promedio evaluaciones mas altas.


```python
sns.barplot(x ='division', y = 'eval', data = ratings_df)
```

Comparad la edad con el tipo de jornada de trabajo y el género, utilizando un gráfico de cajas (boxplot).


```python
sns.boxplot(x = 'tenure', y = 'age', hue = 'gender', data = no_duplicates_ratings_df)
```

Borrad del dataframe original(ratings_df) las columnas sobrantes ya convertidas en variables ficticias (dummies).


```python
ratings_df = ratings_df.drop(columns = ['minority','gender','credits','division','native','tenure'])
ratings_df.head(5)
```

Realizad un agrupamiento sobre la variable edad.

```python
def getAgeGroup(age):
    if np.isnan(age):
        return np.nan
    elif age < 21:
        return 0 # 'Teenager'
    elif age >= 21 and age <=34:
        return 1 # 'YoungAdult'
    elif age >=34 and age<=49:
        return 2 # 'Adult'
    elif age >= 49 and age<=65:
        return 3 # 'OldAdult'
    elif age > 65:
        return 4 # 'Elderly'y
```

```python
ratings_df['age'] = ratings_df['age'].apply(getAgeGroup)
```
