# Cuartiles y  datos atÃ­picos

### Limpieza por cuartiles

Los cuartiles son valores que dividen una muestra de datos en cuatro partes iguales. Con los cuartiles se puede evaluar rÃ¡pidamente la dispersiÃ³n y la tendencia central de los datos.

El recuadro se extiende desde los valores de cuartil Q1 a Q3 de los datos, con una lÃ­nea en la mediana (Q2). Los **outliers** se extienden desde los bordes de la **caja** para mostrar el rango de los datos. 
La posiciÃ³n de los **outliers** se establece de forma predeterminada en 1.5 * IQR (IQR = Q3 - Q1) desde los bordes de la caja.

### Analizando la columna SalePrice

Mostrar un grÃ¡fico boxplot para la columna SalePrice.

```python
import matplotlib.pyplot as plt

df.boxplot('SalePrice')
```

Obtenemos el Rango intercuartil (IQR) y los datos atÃ­picos situados en el lÃ­mite inferior Q1 - 1.5 * IQR y el lÃ­mite superior Q3 + 1.5 * IQR

```python
SalePrice = df['SalePrice']
IQR = 1.5*(SalePrice.quantile(.75)-SalePrice.quantile(.25))
lim_sup = SalePrice.quantile(.75)+IQR
lim_inf = SalePrice.quantile(.25)-IQR
print(lim_sup)
print(lim_inf)
```

```python
SalePrice_clean = SalePrice[(SalePrice>=lim_sup) | (SalePrice<=lim_inf)]
SalePrice_clean
```

Borramos los datos atÃ­picos obtenidos.(No sobreescribir el dataframe)

```python
dfquantil = df
dfquantil.drop(SalePrice_clean.index)
```

### Limpieza por desviaciÃ³n estandar

La desviaciÃ³n estÃ¡ndar es una de las medidas de estadÃ­stica mas populares y de mayor uso. Esta medida indica el grado de dispersiÃ³n alrededor de la media. Una desviaciÃ³n grande indica datos muy dispersos y una desviaciÃ³n pequeÃ±a indica un menor grado de dispersiÃ³n.

ObtÃ©n y borra los outliers utilizando el mÃ©todo de la desviaciÃ³n estÃ¡ndar.

```python
dfstd = df
factor = 3
# upper_lim = dfstd['SalePrice'].mean() + dfstd['SalePrice'].std() * factor
# lower_lim = dfstd['SalePrice'].mean() - dfstd['SalePrice'].std() * factor
# dfstd_clean = (dfstd['SalePrice'] < upper_lim) & (dfstd['SalePrice'] >
# lower_lim)
# dfstd = dfstd[(dfstd['SalePrice'] < upper_lim) & (dfstd['SalePrice'] >
# lower_lim)]

SalePrice = df['SalePrice']
lim_sup = SalePrice.mean() + SalePrice.std() * factor
lim_inf = SalePrice.mean() - SalePrice.std() * factor
sp_std = SalePrice[(SalePrice<lim_inf)|(SalePrice>lim_sup)]
```

```python
dfstd = dfstd.drop(sp_std.index)
```

```python
dfstd
```

Otra manera de obtener los outliers mediante desviaciÃ³n tÃ­pica, es utilizar el puntaje Z (**Z -score**). 
El Z-score es el nÃºmero de desviaciones estÃ¡ndar por las cuales el valor de una observaciÃ³n estÃ¡ por encima del valor medio de lo que se estÃ¡ observando.

Haremos la prueba sobre la misma columna de antes. 

Mostramos el valor Z-score con la librerÃ­a scipy.

```python
from scipy import stats

dfz = df
SalePrice = df["SalePrice"]
z = stats.zscore(SalePrice)
sP_clean = SalePrice[(z>3) | (z<-3)]
```

Limpiamos los datos atÃ­picos que estÃ¡n por encima y por debajo del valor Z-score.

```python
sP_clean
```

```python
dfz = dfz.drop(sP_clean.index)
```
