# Cuartiles y  datos atípicos

### Limpieza por cuartiles

Los cuartiles son valores que dividen una muestra de datos en cuatro partes iguales. Con los cuartiles se puede evaluar rápidamente la dispersión y la tendencia central de los datos.

El recuadro se extiende desde los valores de cuartil Q1 a Q3 de los datos, con una línea en la mediana (Q2). Los **outliers** se extienden desde los bordes de la **caja** para mostrar el rango de los datos. 
La posición de los **outliers** se establece de forma predeterminada en 1.5 * IQR (IQR = Q3 - Q1) desde los bordes de la caja.

### Analizando la columna SalePrice

Mostrar un gráfico boxplot para la columna SalePrice.

```python
import matplotlib.pyplot as plt

df.boxplot('SalePrice')
```

Obtenemos el Rango intercuartil (IQR) y los datos atípicos situados en el límite inferior Q1 - 1.5 * IQR y el límite superior Q3 + 1.5 * IQR

```python
SalePrice = df['SalePrice']
IQR = 1.5*(SalePrice.quantile(.75)-SalePrice.quantile(.25))
lim_sup = SalePrice.quantile(.75)+IQR
lim_inf = SalePrice.quantile(.25)-IQR
print(lim_sup)
print(lim_inf)
```

```python
SalePrice_clean = SalePrice[(SalePrice>=lim_sup) | (SalePrice<=lim_inf)]
SalePrice_clean
```

Borramos los datos atípicos obtenidos.(No sobreescribir el dataframe)

```python
dfquantil = df
dfquantil.drop(SalePrice_clean.index)
```

### Limpieza por desviación estandar

La desviación estándar es una de las medidas de estadística mas populares y de mayor uso. Esta medida indica el grado de dispersión alrededor de la media. Una desviación grande indica datos muy dispersos y una desviación pequeña indica un menor grado de dispersión.

Obtén y borra los outliers utilizando el método de la desviación estándar.

```python
dfstd = df
factor = 3
# upper_lim = dfstd['SalePrice'].mean() + dfstd['SalePrice'].std() * factor
# lower_lim = dfstd['SalePrice'].mean() - dfstd['SalePrice'].std() * factor
# dfstd_clean = (dfstd['SalePrice'] < upper_lim) & (dfstd['SalePrice'] >
# lower_lim)
# dfstd = dfstd[(dfstd['SalePrice'] < upper_lim) & (dfstd['SalePrice'] >
# lower_lim)]

SalePrice = df['SalePrice']
lim_sup = SalePrice.mean() + SalePrice.std() * factor
lim_inf = SalePrice.mean() - SalePrice.std() * factor
sp_std = SalePrice[(SalePrice<lim_inf)|(SalePrice>lim_sup)]
```

```python
dfstd = dfstd.drop(sp_std.index)
```

```python
dfstd
```

Otra manera de obtener los outliers mediante desviación típica, es utilizar el puntaje Z (**Z -score**). 
El Z-score es el número de desviaciones estándar por las cuales el valor de una observación está por encima del valor medio de lo que se está observando.

Haremos la prueba sobre la misma columna de antes. 

Mostramos el valor Z-score con la librería scipy.

```python
from scipy import stats

dfz = df
SalePrice = df["SalePrice"]
z = stats.zscore(SalePrice)
sP_clean = SalePrice[(z>3) | (z<-3)]
```

Limpiamos los datos atípicos que están por encima y por debajo del valor Z-score.

```python
sP_clean
```

```python
dfz = dfz.drop(sP_clean.index)
```
